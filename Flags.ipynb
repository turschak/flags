Import the ipythonblocks package

```python
from ipythonblocks.ipythonblocks import *
```

Set the basic information about your flag: dimensions, default color, and block size

```python
width = 25
height = 13
fill = colors['Gray']
block_size = 20
```

Create the basic "flag" and display it: 

```python
flag = BlockGrid(width, height, fill, block_size)
flag
```

Now, change the colors of blocks to create a world flag.  There are many ways to select which groups of blocks you want to change - indexing, for loops and conditional statements.  

To set the colors, you will either need an RGB tuple: `(0 255 0)` or you can use `ipythonblocks` built-in colors, like so: `colors['ColorName']`.  A full list of built in colors (and what they look like) is available in the other `Colors.ipynb` notebook.  

```python
# red and white stripes (using for loop and conditional statement)
# for loop goes through numbers 0 through height of flag
#for row in range(height):
#    # if row is even (1st through end, make red).  Otherwise, make even
#    if row %2 == 0:
#        flag[row,:] = colors['Red']
#    else: 
#        flag[row,:] = colors['White']

# alternatively, red and white stripe using fancy list indexing
flag[0::2 ,:] = colors['Red']
flag[1::2 ,:] = colors['White']

# blue corner (using simple indexing - easiest option!)
flag[0:6, 0:10] = colors['Blue']

# stars (for loop and fancy list indexing)
# for loop goes through relevant 4 rows of grid
for row in [1,2,3,4]:
    # always need to end before the last blue column
    end = 9
    # if row is even, "offset" the first white square
    if row % 2 == 1:
        start = 1
    else: 
        start = 2
    # for each row in the loop, go from start to end values
    flag[row, start:end:2] = colors['White']

# print flag
flag
```
